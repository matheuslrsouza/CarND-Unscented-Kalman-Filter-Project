```python
import numpy as np
import matplotlib.pyplot as plt
import csv

#rmse = [.17, .18, .76, .32]


file = ('/Users/matheus.souza/desenvolvimento/projetos/sdcnd/' + 
        'term2/CarND-Unscented-Kalman-Filter-Project/data/nis_radar.txt')

plt.figure(figsize=(10,5))

with open(file) as lines:
    nis = [];
    for line in lines:
        nis.append(line)
    
    x = np.arange(len(nis))
    plt.plot(x, nis)
    
    plt.axhline(y=7.8, color='r', linestyle='-')


    plt.xlabel('iteration')
    plt.ylabel('NIS value')
    plt.title('NIS RADAR')
    plt.grid(True)
    plt.savefig("radar_nis.png")

    plt.show()
```

```python
# init comparison with ground truth values

file = ('/Users/matheus.souza/desenvolvimento/projetos/sdcnd/' + 
        'term2/CarND-Unscented-Kalman-Filter-Project/data/data_gt_and_pred.txt')

gt_px = []
pred_px = []
gt_py = []
pred_py = []
gt_vx = []
pred_vx = []
gt_vy = []
pred_vy = []
gt_yaw = []
pred_yaw = []
gt_yaw_rate = []
pred_yaw_rate = []

with open(file) as lines:
    for line in lines:
        data = line.split(" ")
        
        gt_px.append(data[0])
        pred_px.append(data[1])        
        gt_py.append(data[2])
        pred_py.append(data[3])
        gt_vx.append(data[4])
        pred_vx.append(data[5])
        gt_vy.append(data[6])
        pred_vy.append(data[7])
        gt_yaw.append(data[8])
        pred_yaw.append(data[9])
        gt_yaw_rate.append(data[10])
        pred_yaw_rate.append(data[11])
```

```python
plt.figure(figsize=(10,4))
      
x = np.arange(len(gt_px))

plt.xlabel('iteration')
plt.ylabel('px')
plt.title('Position X')

plt.plot(x, pred_px, label="predicted")
plt.plot(x, gt_px, label="ground truth")

# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.grid(True)
plt.savefig("px.png")

plt.show()
```
